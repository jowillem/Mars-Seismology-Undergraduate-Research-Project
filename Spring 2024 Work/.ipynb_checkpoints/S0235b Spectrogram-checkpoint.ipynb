```python
import matplotlib.pyplot as plt
import numpy as np
from obspy import read
from scipy.signal import spectrogram
from obspy import UTCDateTime
```

```python
stream = read(r'C:\Users\wille\Mars-Seismology-Undergraduate-Research-Project\InSight-seismic-data-downloader\DATA\BROADBAND\A\S0235b\S0235b_DISP.mseed')

data = stream[0].data
sampling_freq = stream[0].stats.sampling_rate
```

```python
# Set parameters for the spectrogram
nperseg = 256  # Number of data points in each segment
noverlap = 128  # Number of overlapping points between segments

# Create the spectrogram
frequencies, times, Sxx = spectrogram(data, fs=sampling_freq, nperseg=256*4) #nperseg samples per segment, balance between high resolution and low resolution.
times -= 60*30
print(Sxx.shape)
print(frequencies.shape, len(frequencies))

# Plot the spectrogram
plt.figure(figsize=(12, 6))
plt.pcolormesh(times, frequencies, 10 * np.log10(Sxx), shading='gouraud', cmap="plasma", vmin=-240, vmax=-170)
plt.title('Spectrogram of s0235b Marsquake')
plt.xlabel('Time (s)')
plt.ylabel('Frequency (Hz)')
plt.colorbar(label='Intensity (dB)')
plt.axvline(2100)
plt.ylim(0.05, 1)  # Limit the y-axis to focus on lower frequencies
plt.xlim(0,2000)
plt.show()
```
