```python
import matplotlib.pyplot as plt
import numpy as np
from obspy import read
from scipy.signal import spectrogram
from obspy.clients.fdsn import Client
from obspy import UTCDateTime
```

```python
stream = read(r'C:\Users\wille\Mars-Seismology-Undergraduate-Research-Project\InSight-seismic-data-downloader\DATA\LOW_FREQUENCY\A\S0173a\S0173a_DISP.mseed')

data = stream[0].data
sampling_freq = stream[0].stats.sampling_rate

# Set parameters for the spectrogram
nperseg = 256  # Number of data points in each segment
noverlap = 128  # Number of overlapping points between segments

# Create the spectrogram
frequencies, times, Sxx = spectrogram(data, fs=sampling_freq, nperseg=256*4) #nperseg samples per segment, balance between high resolution and low resolution.
times -= 60*30

# Plot the spectrogram
plt.figure(figsize=(12, 6))
plt.pcolormesh(times, frequencies, 10 * np.log10(Sxx), shading='gouraud', cmap="plasma", vmin=-200, vmax=-160)
plt.title('S0173a Spectrogram')
plt.xlabel('Time (s)')
plt.ylabel('Frequency (Hz)')
plt.colorbar(label='Intensity (dB)')
plt.ylim(0.1, 1)  # Limit the y-axis to focus on lower frequencies
plt.xlim(0,750)
plt.show()

stcopy = stream.copy()
stcopy.taper(0.05)
stcopy.filter("bandpass", freqmin=0.2, freqmax=0.9, corners=4)
stcopy[0].plot();

trz = stcopy.select(channel="BHZ")[0]
origin_time = 60*30
p_predicted = 226.497
s_predicted = 403.212
time_for_trz = trz.times() - origin_time

plt.figure(figsize=(6, 4))
plt.plot(time_for_trz, trz.data)
plt.xlim(0,750)
plt.ylim(-5e-9,5e-9);
```
